```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('KMeans.csv')
df
```

```python
X=df.iloc[:,[1,2]].values
X
```

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customer') 
plt.ylabel('Distance')
plt.show()
```
